```python
#Write your code here

# TODO
class Shape:
    def __init__(self, shape, size1, size2):
        self.shape = shape
        self.size1 = size1
        self.size2 = size2
    def area(self):
        if self.shape == 'Triangle':
            self.area = .5 * self.size1 * self.size2
        elif self.shape == "Square" or self.shape == "Rectangle":
            self.area = self.size1 * self.size2
        elif self.shape == "Rhombus":
            self.area = .5 * self.size1 * self.size2
        else:
            self.area = 'Shape unknown'
        print("Area:", self.area)
# Do not change the following code:
triangle_1 = Shape("Triangle",10,25)
triangle_1.area()
print("==========================")
square_1 = Shape("Square",10,10)
square_1.area()
print("==========================")
rhombus_1 = Shape("Rhombus",18,25)
rhombus_1.area()
print("==========================")
rectangle_1 = Shape("Rectangle",15,30)
rectangle_1.area()
print("==========================")
trapezium_1 = Shape("Trapezium",15,30)
trapezium_1.area()
```

```python
#Write your code here
# TODO
class Calculator:
    def __init__(self, num1, num2, op):
        self.num1 = num1
        self.num2 = num2
        self.op = op
    def calculate_button(self):
        if self.op == '+':
            self.val = self.num1 + self.num2
        elif self.op == '-':
            self.val = self.num1 - self.num2
        elif self.op == '*':
            self.val = self.num1 * self.num2
        elif self.op == '/':
            self.val = self.num1 / self.num2
        return self.val
    def details(self):
        print(f' {self.num1} {self.op} {self.num2} = {self.val}')

#Do not change the following code:
add = Calculator(10, 20, '+')
val = add.calculate_button()
print("Returned value:", val)
add.details()
print("==================")

sub = Calculator(val, 10, '-')
val = sub.calculate_button()
print("Returned value:", val)
sub.details()
print("==================")

div = Calculator(val, 5, '*')
val = div.calculate_button()
print("Returned value:", val)
div.details()
print("==================")

mult = Calculator(val, 16, '/')
val = mult.calculate_button()
print("Returned value:", val)
mult.details()
print("==================")
```
